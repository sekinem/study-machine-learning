```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data[50:, 2:]
y = iris.target[50:] - 1
```

```python
X
```

```python
y
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['versicolor', 'virginica'], loc='best')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
svm = LinearSVC().fit(X_train, y_train)

def plot_separator(model):
    plt.figure(figsize=(10, 6))
    mglearn.plots.plot_2d_separator(model, X)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
    plt.xlabel('petal length')
    plt.ylabel('petal width')
    plt.xlim(2.8, 7.0)
    plt.ylim(0.8, 2.6)
    plt.show()

plot_separator(svm)
```

```python
svm100 = LinearSVC(C=100).fit(X_train, y_train)
plot_separator(svm100)
```

```python
print('score on training sets: {:.2f}'.format(svm.score(X_train, y_train)))
print('score on test sets: {:.2f}'.format(svm.score(X_test, y_test)))
```

```python
print('score on training sets: {:.2f}'.format(svm100.score(X_train, y_train)))
print('score on test sets: {:.2f}'.format(svm100.score(X_test, y_test)))
```

```python
svm001 = LinearSVC(C=0.01).fit(X_train, y_train)
plot_separator(svm001)
```

```python
print('score on training sets: {:.2f}'.format(svm001.score(X_train, y_train)))
print('score on test sets: {:.2f}'.format(svm001.score(X_test, y_test)))
```
