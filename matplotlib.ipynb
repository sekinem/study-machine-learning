```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)

import seaborn as sns
sns.set_style('darkgrid')
```

```python
x = np.linspace(0, 1, 10)
x
```

```python
y = 2 * x + 0.1 * np.random.randn(10)
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
```

```python
x1 = np.linspace(0, 1, 10)
y1 = x1 + 0.1 * np.random.randn(10)
x2 = np.linspace(0, 1, 100)
y2 = 2 * x2 + 0.1 * np.random.randn(100)
x3 = np.linspace(0, 1, 1000)
y3 = 3 * x3 + 0.1 * np.random.randn(1000)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x3, y3)

plt.show()
```

```python
fig = plt.figure(figsize=(20, 12))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x3, y3)

plt.show()
```

```python
df = pd.DataFrame({
    'C1': ['A', 'A', 'A', 'B', 'B', 'C'],
    'C2': [20, 50, 60, 80, 100, 30],
    'C3': [40, 200, 100, 500, 40, 200],
})
df
```

```python
df = pd.DataFrame(
    data=100 * np.random.rand(5, 5),
    index=['A', 'B', 'C', 'D', 'E'], 
    columns=['C1', 'C2', 'C3', 'C4', 'C5']
)
df.head()
```

```python
plt.figure(figsize=(12, 8))

x = np.arange(len(df))

bar_width = 0.25

plt.bar(x, df['C1'], color='red', width=bar_width, label='C1')
plt.bar(x + bar_width, df['C2'], color='blue', width=bar_width, label='C2')

plt.xticks(x + bar_width / 2, df.index, fontsize=20)
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(16, 8))

x = np.arange(len(df), dtype='float64')

bar_width = 0.1

for col in df.columns:
    plt.bar(x, df[col], width=bar_width, label=col)
    x += bar_width
    
plt.xticks((np.arange(len(df)) + x - bar_width) / 2, df.index, fontsize=20)
plt.legend()

plt.show()
```

```python
df = pd.DataFrame(data=np.random.randint(0, 5, 500), columns=['C1'])
df.head()
```

```python
plt.figure(figsize=(10, 10))

textprops = {'fontsize': 20}

plt.pie(
    df['C1'].value_counts(), 
    labels=df['C1'].unique(),
    autopct='%.1f%%',
    startangle=90,
    counterclock=False,
    textprops=textprops
)

plt.show()
```
