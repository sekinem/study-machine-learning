```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv('kaggle_data/titanic/train.csv')
test_df = pd.read_csv('kaggle_data/titanic/test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
all_df = pd.concat((train_df.loc[:, 'Pclass':'Embarked'], test_df.loc[:, 'Pclass':'Embarked']))
```

```python
all_df.info()
```

```python
all_df['Age'] = all_df['Age'].fillna(all_df['Age'].mean())
all_df['Fare'] = all_df['Fare'].fillna(all_df['Fare'].mean())
all_df['Embarked'] = all_df['Embarked'].fillna(all_df['Embarked'].mode()[0])
```

```python
all_df.info()
```

```python
all_df.head()
```

```python
cat_features = ['Sex', 'Embarked']

for col in cat_features:
    lbl = LabelEncoder()
    all_df[col] = lbl.fit_transform(list(all_df[col].values))
```

```python
all_df.head()
```

```python
all_df = all_df.drop(columns=['Name', 'Ticket', 'Cabin'])
```

```python
all_df.head()
```

```python
train = all_df[:train_df.shape[0]]
test = all_df[train_df.shape[0]:]
```

```python
train.info()
```

```python
y = train_df['Survived']
ID = test_df['PassengerId']
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=0)
```

```python
import xgboost as xgb
```

```python
params = {
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'eta': 0.1,
    'max_depth': 6,
    'subsample': 1,
    'colsample_bytree': 1,
}

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

model = xgb.train(
    params=params,
    dtrain=dtrain,
    num_boost_round=100,
    early_stopping_rounds=10,
    evals=[(dtest, 'test')]
)
```

```python
prediction = model.predict(xgb.DMatrix(X_test), ntree_limit=model.best_ntree_limit)
prediction
```

```python
prediction = np.where(prediction < 0.5, 0, 1)
prediction
```
